```python
###Example 1: A user has provided the string l vey u and the character o, and the output will be loveyou.

inp = "l vey u"
print(inp.replace(" ", "o"))

def removeDuplicates(array):
    size = len(array)
    if size == 0:
        return 0
    insertIndex = [array[0]]
    for i in range(1, size):
        if array[i] != insertIndex[-1]:
            insertIndex.append(array[i])
      
    return len(insertIndex)

array_1 = [1, 2, 2, 3, 3, 4]
k1 = removeDuplicates(array_1)
print(k1)
# 4; array_1[:k1] -> [1, 2, 3, 4]

array_2 = [1, 1, 3, 4, 5, 6, 6]
k2 = removeDuplicates(array_2)
print(k2)
# 5; array_2[:k2] -> [1, 3, 4, 5, 6]


def is_palindrome(s):
    s = ''.join(e for e in s if e.isalnum()).lower()  # Remove non-alphanumeric and convert to lowercase
    return s == s[::-1]

print(is_palindrome("A man, a plan, a canal: Panama"))  # Output: True
```

```python
#Can you find the maximum single sell profit
def buy_sell_stock_prices(prices):
    curr_buy = prices[0]
    curr_sell = prices[1]
    max_profit = curr_sell - curr_buy

    for i in range(1,len(prices)):
        profit = prices[i] - curr_buy
        if profit > max_profit:
            max_profit = profit
            curr_sell = prices[i]
        if prices[i] < curr_buy:
            curr_buy = prices[i]

    return curr_buy,curr_sell

stock_prices_1 = [10,9,16,17,19,23]
print(buy_sell_stock_prices(stock_prices_1))
stock_prices_2 = [8, 6, 5, 4, 3, 2, 1]
print(buy_sell_stock_prices(stock_prices_2))

```

```python
#Can you find a Pythagorean triplet in an array?
def checkTriplet(arr):
    squares = [x*x for x in arr]
    squares.sort()
    for i in range(0,len(squares)-2):
        if squares[i] + squares[i+1] == squares[i+2]:
            return True
    return False


arr = [3, 2, 4, 6, 5]
checkTriplet(arr)
# True
```
